```python
import pandas as pd
```

```python
df = pd.read_csv('../data/mpst_full_data.csv')
```

```python
df['tags']
```

```python
for row in df['tags']:
    print(row)
```

```python
tags = []

def split_tags(row):
    row = row.split(',')
    return row
            
for row in df['tags']:
    split = split_tags(row)
    for tag in split:
        if tag.strip() not in tags:
            tags.append(tag.strip())
            
print(tags)

with open('../data/tags_list.txt', 'w') as f:
    f.write(','.join(tags))
```

```python
with open('../data/tags_list.txt', 'r') as f:
    s = f.read()
    tags = s.split(',')
    print(tags)
    print(len(tags))
```

```python
max_chars_per_column = {}

for column in df.columns:
    if df[column].dtype == 'object':  # Check if the column contains strings
        max_chars_per_column[column] = df[column].str.len().max()

max_chars_overall = max(max_chars_per_column.values())

print("Maximum number of characters overall:", max_chars_overall)
print("Maximum number of characters per column:", max_chars_per_column)

```
